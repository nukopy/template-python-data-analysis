```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

import japanize_matplotlib
plt.style.use('ggplot') # ref: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
```

```python
import pathlib

DIRNAME = "basic_matplotlib"
OUTPUT_DIR = f"../../output/notebooks/{DIRNAME}"

# mkdir
p = pathlib.Path(OUTPUT_DIR)  # .resolve()
p.mkdir(parents=True, exist_ok=True)
```

```python
# 
x1 = [1, 2, 3, 4, 5]
y1_list = [
    [2, 5, 3, 3, 4],
    [1, 2, 1, 2, 5],
]

x2 = np.arange(-3, 3, 0.01)
y2_list = [
    np.sin(x2),
    np.cos(x2),
]

# 
nrows = 2
ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols) # axes  ndarray
fig.tight_layout()

# ax1 = axes[]
# ax2 = fig.add_subplot(212)
for row in range(nrows):
    x: list
    y: list

    for col in range(ncols):
        # x 
        if col == 0:
            x = x1
        if row == 1:
            x = x2

        # y 
        if row == 0:
            y = y1_list[col]
        elif row == 1:
            y = y2_list[col]
        
        # ax 
        ax = axes[row][col]

        # 
        ax.plot(x, y)

        # 
        ax.set_title(f" ({row}, {col})")

# save figure
fname = f"{OUTPUT_DIR}/line-plot.png"
fig.tight_layout()
fig.savefig(fname=fname)
```

```python
# create graph
G = nx.petersen_graph()

# plot
nrows = 1
ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols) # axes  ndarray
fig.tight_layout()

ax1 = axes[0]
nx.draw(G, ax=ax1, with_labels=True, font_weight='bold')

ax2 = axes[1]
nx.draw_shell(G, ax=ax2, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')

# save figure
fname = f"{OUTPUT_DIR}/network-plot.png"
fig.savefig(fname=fname)
```
